VOLATILTY ANALYSIS

Moving averages smooth out price data to identify trends by averaging values over a specified period.Mostly used to indentify flactuations in price over a soecified period of time.

```python
Daily_data['Moving Average[30]'] = Daily_data['Close'].rolling(window = 30).mean()
Daily_data['Moving Average[30]'].head(20)

Daily_data['Moving Average[30]'] = Daily_data['Open'].rolling(window = 30).mean()
Daily_data['Moving Average[30]'].head(20)

Daily_data[0:20]
```

PLOT FOR THE MOVING AVERAGES

```python
plt.figure(figsize=(20,20))
sns.lineplot(x = 'Date', y = 'Moving Average[30]', data = Daily_data)
plt.xlabel("Date")
plt.ylabel('Running Averages')
plt.title("Moving averages per day")
plt.xticks(rotation = 90)
plt.show()
```

VOLATILITY INDEX IS THE STANDARD DEVIATION OF THE PRICES OVER TIME SHOWING THE VARIABILITY OF THE PRICES OVER SAID TIME

```python
Daily_data['Volatility'] = Daily_data['Close'].rolling(window = 30).std()
Daily_data['Volatility'].head(20)

Daily_data['Volatility'] = Daily_data['Open'].rolling(window = 30).std()
Daily_data['Volatility'].head(20)
```

```python
plt.figure(figsize=(20,20))
sns.lineplot(x = 'Date', y = 'Volatility', data = Daily_data, color = 'red')
plt.xlabel("Date")
plt.ylabel('Volatility')
plt.title("Variability of the prices per day")
plt.xticks(rotation = 90)
plt.show()
```
